```python
N=10000
p=np.array([0., .3,.1,.6,])[:-1, None]
cum_p = np.cumsum(p, axis=0)
print(cum_p.shape)
rnd_sample = np.random.rand(1, N)
print(rnd_sample)
samples = (rnd_sample > cum_p)
print(samples.shape)
samples=samples.sum(axis=0)
print(samples.shape)
np.unique(samples, return_counts=True)
```
